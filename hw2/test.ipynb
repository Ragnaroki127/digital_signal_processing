```python
import numpy as np
import matplotlib.pyplot as plt
from DFT import generate_WN
%matplotlib inline
```

### Test for exercise 2

```python
N1 = 100
x1 = np.arange(0, 5, 5 / N1)
x1_n = np.array([1] * 20 + [0] * 80)
plt.plot(x1, x1_n, 'r-')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Positive half of the signal')
plt.show()
```

```python
WN_1 = generate_WN(N1)
X1_k = WN_1 @ x1_n[np.newaxis].T
X1 = np.arange(0, 2 * np.pi, 2 * np.pi / N1)
plt.plot(X1, np.abs(X1_k.T.squeeze()), 'b-')
plt.xlabel('frequency')
plt.ylabel('amplitude')
plt.title('Frequency domain of the positive half of the signal')
plt.show()
```

```python
N2 = 200
x2 = np.arange(0, 10, 10 / N2)
x2_n = np.array([1] * 40 + [0] * 160)
plt.plot(x2, x2_n, 'r-')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Signal amplitude after shifting')
plt.show()
```

```python
WN_2 = generate_WN(N2)
X2_k = WN_2 @ x2_n[np.newaxis].T
X2 = np.arange(0, 2 * np.pi, 2 * np.pi / N2)
plt.plot(X2, np.abs(X2_k.T.squeeze()), 'b-')
plt.xlabel('frequency')
plt.ylabel('amplitude')
plt.title('Frequency domain of the signal after shifting')
plt.show()
```
